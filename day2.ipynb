**operators**
1. arith -> + , _ , * , / , // , % , **

2. relational -> > , < , >= , <== , == , !=

3. assignment -> += , -=, *= , %+

4. logical -> and , or , not

5. bitwise -> & , |

```python
# arithmetic opr
num1 = 10
num2 = 20
print(num1 + num2) # 30
```
