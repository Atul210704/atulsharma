```python
marks=int(input("Enter marks: "))
if marks>=95:
    print("Grade A")
elif marks>=90:
    print("Grade B")
elif marks>=75:
    print("Grade C")
elif marks>=60:
    print("Grade D")
else:
    print("Fail")
```

```python
#list
l = [10,20,30]
#positive indexing
print(l[0])
print(l[1])
print(l[2])
#negative indexing
print(l[-1])
print(l[-2])
print(l[-3])

# mutable
l[0] = 100
l[-1] = 200
print(l)
```
